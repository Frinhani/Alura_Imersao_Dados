###Desafio 06: Estudar o código que plota a matriz de correlação (heatmap)

A documentação para o heatmap https://seaborn.pydata.org/generated/seaborn.heatmap.html 

````python

mask = np.triu(np.ones_like(corr_celular,dtype=bool))
# A máscara do triangulo superior: O np.triu pega o triangulo superior do Array,
# o np.ones_like retorna um array com a informação do dtype, nesse caso boleano

f, ax = plt.subplots(figsize=(16, 14))
#Cria a área onde o heatmap será plotado

cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Configuração da paleta de cores, usando duas cores, 
#o as_cmap serve para mostrar o mapa de cor

sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
# Cria o mapa de calor, passando os dados,
#mask = a máscara não mostra as células onde o resultado for True
#cmap = é a nossa configuração de paleta de cores
#center = o centro do mapa de cor quando temos informações divergentes
#square = para que a células fiquem quadradas
#linewidths = o tamanho das linhas que dividem as células
#cbar_kws= para definir a cor das barras
````
